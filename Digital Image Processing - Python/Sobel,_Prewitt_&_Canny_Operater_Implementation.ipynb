```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

 
 
def edge_detection(image, filter):
    new_image_x = cv2.filter2D(image, -1,filter)    # horizontal edge
    new_image_y = cv2.filter2D(image, -1,np.flip(filter.T, axis=0)) # vertical egde

    grad = np.hypot(new_image_x,new_image_y) # Gradient calculated
    grad = grad.astype('float64') 
  
    plt.imshow(grad, cmap='gray')
    plt.title("Gradient Magnitude")
    plt.show()
 
 
filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) # sobel filter
filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]) # prewitt filter
image = cv2.imread('building.tif',0)
print("Sobel Operator")
edge_detection(image, filter)
print("Prewitt Operator")
edge_detection(image, filter)
image = cv2.GaussianBlur(image,(15,15),0)
print("Sobel Operator")
edge_detection(image, filter)
print("Prewitt Operator")
edge_detection(image, filter)

# sobel operator is a derivate mask and coefficent of the masks aren't fixed unlike prewitt operator. There is hardly any difference in the ouput image, the only diffrernce in our case are the filters for both are different.
```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

 
def edge_detection(image, filter):
    new_image_x = cv2.filter2D(image, -1,filter) # horizontal egde
    new_image_y = cv2.filter2D(image, -1,np.flip(filter.T, axis=0)) # vertical egde

    grad = np.hypot(new_image_x,new_image_y)/255 # gradient calculated
    grad = grad.astype('float64') 
  
    plt.imshow(grad, cmap='gray')
    plt.title("Gradient Magnitude")
    plt.show()
 
 
filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])  # sobel filter
filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])  # prewitt filter

image = cv2.imread('CT.tif',0)
img = cv2.GaussianBlur(image,(15,15),0)

print("Sobel Operator")
edge_detection(img, filter)

print("Prewitt Operator")
edge_detection(img, filter)

print("Canny Operator")
cannyImg = cv2.Canny(image,240,250)  # canny operator used with min threshold = 240 and max threshold = 250
plt.imshow(cannyImg, cmap='gray')
plt.title("Combined Edges")
plt.show()

```

```

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('Fig01.tif',0)

ellipseMask = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
kernel = np.ones((3,3),np.uint8)
erosion = cv2.dilate(img,ellipseMask,iterations = 1)
kernel2 = np.ones((7,7),np.uint8)
erosion2 = cv2.dilate(img,kernel2,iterations = 1)

cv2_imshow(img)
print("\n\nDilation")
print("\n\nKernel: (3,3)")
cv2_imshow(erosion) 
print("\n\nKernel: (7,7)")
cv2_imshow(erosion2) 
```

```

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('Fig01.tif',0)

ellipseMask = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, ellipseMask)

cv2_imshow(img)
print("\n\nDilating then Eroding")
cv2_imshow(closing) 
```

```

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import skimage.morphology

img = cv2.imread('Objects.png',0)

rectmask = cv2.getStructuringElement(cv2.MORPH_RECT, (71, 71))
crossmask = skimage.morphology.selem.diamond(49, np.uint8)
ellipsemask = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (77, 77))
opening1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, rectmask)
opening2 = cv2.morphologyEx(img, cv2.MORPH_OPEN, crossmask)
opening3 = cv2.morphologyEx(img, cv2.MORPH_OPEN, ellipsemask)

cv2_imshow(img)

print("\n\nOpening")
print("\n\nRectangular Object:")
cv2_imshow(opening1) 
print("\n\nDiamond Object:")
cv2_imshow(opening2) 
print("\n\nCircular Object:")
cv2_imshow(opening3) 
```

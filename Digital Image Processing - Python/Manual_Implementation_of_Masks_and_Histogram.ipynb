```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('kidney.tif',0)

cv2_imshow(img)
x,y = img.shape
img1 = np.zeros([x,y])
mask = np.zeros([x,y])
print(x,y)
for i in range(0,x-1):
  for j in range(0,y-1):
    if img[i,j]>150:
      mask[i,j] = 255
    else:
      mask[i,j] = img[i,j]

for i in range(0,x-1):
  for j in range(0,y-1):
    img1[i,j] = img[i,j] and mask[i,j]
      

cv2_imshow(img1)

for i in range(0,x-1):
  for j in range(0,y-1):
    if img[i,j]>150:
      mask[i,j] = 255
    else:
      mask[i,j] = 0

for i in range(0,x-1):
  for j in range(0,y-1):
    img1[i,j] = img[i,j] and mask[i,j]

cv2_imshow(img1)




```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt 

img = cv2.imread('wiki.jpg',0)

cv2_imshow(img)
x,y = img.shape
img1 = np.zeros([x,y])
mask = np.zeros([x,y])
Smax = 255
Smin = 0
Rmax = np.amax(img)
Rmin = np.amin(img)
S = np.zeros([Rmax])
print(x,y)
hist = np.zeros([256])

for i in range(0,x-1):
  for j in range(0,y-1):
    hist[img[i,j]]+=1

plt.plot(hist)
plt.xlim([0,255])
plt.show()

for r in range(Rmin,Rmax):
  S[r] = ((Smax-Smin)/(Rmax-Rmin))*(r-Rmin) + Smin

for i in range(0,x-1):
  for j in range(0,y-1):
    for r in range(Rmin,Rmax):
        if img[i,j] == r:
          img[i,j] = S[r]
      
cv2_imshow(img)

hist = np.zeros([256])

for i in range(0,x-1):
  for j in range(0,y-1):
    hist[img[i,j]]+=1

plt.plot(hist)
plt.xlim([0,255])
plt.show()
```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt 

img = cv2.imread('bright.tif',0)

x,y = img.shape
img1 = np.zeros([x,y])
mask = np.zeros([x,y])
print(x,y)
hist = np.zeros([256])
hist1 = np.zeros([256])
newhist = np.zeros([256])
for i in range(0,x-1):
  for j in range(0,y-1):
    hist[img[i,j]]+=1

plt.plot(hist)
plt.xlim([0,255])
plt.show()
pix = 0
for i in range(0,255):
  pix += hist[i]
  newhist[i] = ((255)/(x*y))*pix

for i in range(0,255):
  hist1[int(newhist[i])] += hist[i]

plt.plot(hist1)
plt.xlim([0,255])
plt.show()
```
